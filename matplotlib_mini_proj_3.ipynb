```python
import pandas as pd
import matplotlib.pyplot as plt



movie = pd.read_csv("/content/movies.csv")
movies = movie.dropna()

plt.figure(figsize=(20,20))
plt.hist(movies['vote_average'], bins=20, color = "blue")
plt.title('Movie Ratings')
plt.xlabel('Ratings')
plt.ylabel('No of ratings')
plt.show()

plt.figure(figsize=(10,10))
movies['revenue'] = movies['vote_count'] * movies['vote_average']
plt.scatter(movies['vote_average'],movies["revenue"])
plt.title('Scatter Plot of Movie Ratings vs revenue')
plt.xlabel('Rating')
plt.ylabel('revenue')


movies['release_date'] = pd.to_datetime(movies['release_date'])
movies['year'] = movies['release_date'].dt.year
movies_per_year = movies['year'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
plt.plot(movies_per_year.index, movies_per_year.values)
plt.title('Trend of Movie Releases Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Movies Released')
plt.show()
```
