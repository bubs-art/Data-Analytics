```python
#pandas mini proj 4
'''Dataset: COVID-19 Cases dataset (CSV)
Tasks:
- Load and clean the data
- Find total cases, recoveries, and deaths
- Identify top affected countries
- Analyze daily growth rate of cases'''
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np



cwl = pd.read_csv("/country_wise_latest.csv")

n = cwl.iloc[:,1:14]
m = n.replace(0, np.nan, inplace=True)
o = n.dropna()


tc = cwl["Confirmed"].sum()
rc = cwl["Recovered"].sum()
de = cwl["Deaths"].sum()

sort = cwl.sort_values(by = "Confirmed",ascending=False)
ac = sort.iloc[:,0:2]

gr = cwl["New cases"]/cwl["Confirmed"]


print(f"new data: {o}")
print("\n\n")
print(f" total cases are: {tc}")
print("\n\n")
print(f"total recovreed cases are: {rc}")
print("\n\n")
print(f"total death occured: {de}")
print("\n\n")
print(f"top affected countries: {ac}")
print("\n\n")
print(f"growth rate: {gr}")
```
