```python
#Numpy mini proj 4
'''Tasks:
- Generate a NumPy array of 100 values representing simulated daily stock prices starting from â‚¹1000.
- Use a random walk model where each day's price is the previous day's price plus a small random change
 (between -â‚¹10 and +â‚¹10).
- Find the maximum, minimum, and average stock prices over the period.
- Identify days when the stock price dropped by more than â‚¹5 compared to the previous day.
- Compute the 5-day moving average of the stock prices.
ðŸ”¹ Concepts Tested: Random number generation, cumulative sums, logical indexing, and moving averages.'''
```

```python
import numpy as np
c = np.random.uniform(-10,10,size = 100)
ip = 1000
sp = ip + (c)


max = np.max(sp); min = np.min(sp); avg = np.average(sp)


diff = np.diff(sp)
y = diff[diff<-5]
z = y.size


s = sp[:5]
a = np.average(s)

print(f"maximum of the dataset: {max}")
print(f"minimum of the dataset: {min}")
print(f"average of the dataset: {max}")
print(f"days where the stock price dropped by more than 5 compared to prev day: {z}")
print(f"5-day average: {a}")
```
