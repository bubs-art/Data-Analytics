```python
#NumPy mini proj 1
'''Objective: Assess understanding of array creation, manipulation, and matrix operations
Tasks:
- Create a 5Ã—5 NumPy array filled with random integers between 1 and 50.âœ…
- Replace all even numbers with -1 without using loops.âœ…
- Compute the sum of each row and store it in a new array.âœ…
- Compute the dot product of the matrix with its transpose.âœ…
- Find the index of the maximum element in the matrix.âœ…
ðŸ”¹ Concepts Tested: Array manipulation, indexing, broadcasting, and matrix operations'''
```

```python
import numpy as np

m = np.random.randint(1,50,size = (5,5))

m[m%2==0] = -1

sum = np.sum(m,axis = 1)

dot = np.dot(m,m.T)

max = np.argmax(m)

print(f"even numbers = -1: {m}")
print(f"Sum of each row is stored in {sum}")
print(f"dot product is {dot}")
print(f"Index of the maximum element is {max}")



```
