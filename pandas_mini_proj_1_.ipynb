```python
#Customer Purchase Analysis
'''Tasks:
- Load the dataset and clean missing values
- Analyze customer purchase behavior (most purchased items, highest spending customers)
- Calculate average order value, top categories
- Find seasonal trends in purchases'''
```

```python
import pandas as pd
import numpy as np
ecdcr = pd.read_csv("/content/ecommerce_customer_data_custom_ratios.csv")

n = ecdcr.iloc[:,12]
m = n.replace(0, np.nan, inplace=True)
k = ecdcr.iloc[:,10]
l = n.replace(0, np.nan, inplace=True)
drop = ecdcr.dropna()

prod = ecdcr["Product Category"].value_counts().sort_index()


sort = ecdcr.loc[ecdcr["Total Purchase Amount"].sort_values(ascending=False),["Customer Name","Total Purchase Amount"]]
svc = sort.value_counts()
vc = svc.sort_index(ascending=True)

am = ecdcr["Total Purchase Amount"].mean()
cat = ecdcr["Product Category"].value_counts()

tp = cat.sum()
tq = ecdcr["Quantity"].sum()
tp = ecdcr["Total Purchase Amount"].sum()

print(f"Clean dataset is \n {drop}")
print("\n\n")
print(f"most purchased items: \n {prod}")
print(f"highest spending customer: {sort}")
print("\n\n")
print(f"Average order value: \n {am}")
print(f"top categories are: \n {cat}")
print(f"\n\n")
print(f"seasonal trends are given below:")
print(f"Total items sold: {tp}")
print(f"Total quantities sold: {tq}")
print(f"Total amount: {tp}")
```
