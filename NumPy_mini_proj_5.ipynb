```python
#numpy proj 3
'''Data Analysis on a NumPy Dataset
Objective: Assess understanding of loading, analyzing, and performing statistical operations on data
Tasks:
â— Generate a NumPy array of 1000 random integers between 10 and 100. âœ…
â— Find the mean, median, standard deviation, and variance of the dataset.âœ…
â— Identify and count how many numbers are greater than the mean.âœ…
â— Normalize the dataset using min-max normalization. âœ…
â— Reshape the array into a 20Ã—50 matrix and find the sum of each column.âœ…
ðŸ”¹ Concepts Tested: Statistical functions, logical indexing, reshaping, and normalization.'''
```

```python
import numpy as np

Ds= np.random.randint(10,101,size = 1000)

mean = np.mean(Ds)
median = np.median(Ds)
stddev = np.std(Ds)
var = np.var(Ds)

gm = Ds[Ds>mean]
count = gm.size

maxval = np.max(Ds)
minval = np.min(Ds)
sqrt = np.sqrt(stddev)
norm = Ds-mean/sqrt


reshape = Ds.reshape(20,50)
sum = np.sum(reshape)

print(f"mean of the dataset is {mean}")
print(f"median of the dataset is {median}")
print(f"standard variation of the dataset is {stddev}")
print(f"variation of the dataset is {var}")
print("---------")
print(f"numbers greater than the mean in the dataset are {gm}")
print(f"count of those numbers are {count}")
print("---------")
print(f"Normalized dataset with the help of max and min is {norm}")
print("---------")
print(f"reshaped dataset in terms of 20x50 matrix is {reshape}")
print(f"sum of the columns in that reshaped dataset is {sum}")




```
