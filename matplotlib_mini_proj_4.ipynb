```python
import pandas as pd
import matplotlib.pyplot as plt
a = pd.read_csv("/content/athlete_events.csv")
ae = a.dropna()

tg = ae.groupby('NOC')["Medal"].value_counts()
tg

sports_medals = ae.groupby('Sport').sum()
sports_medals['Total'] = sports_medals['Gold'] + sports_medals['Silver'] + sports_medals['Bronze']
plt.figure(figsize=(10, 7))
plt.pie(sports_medals.values,labels=["Gold","Silver","Bronze"],autopct='%1.1f%%', startangle=45)
plt.title('Percentage of Total Medals Won in Different Sports')
plt.show()
```
