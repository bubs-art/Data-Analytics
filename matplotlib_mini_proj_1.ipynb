```python
#Sales Data Visualization
'''Tasks:
- Create a bar chart for top-selling product categories.
- Use a pie chart to show the percentage of total sales by region.'''
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

ss = pd.read_csv("/content/SampleSuperstore.csv")



tc = ss.groupby("Category")["Sales"].sum()
tc.plot(kind = "bar")
plt.xlabel = "Category"
plt.ylabel = "Sales sum"
plt.title = "Sales category wise"
plt.show()




sizes = ss.groupby('Region')["Sales"].sum()
labels = ss['Region'].unique()
plt.pie(sizes,labels=labels,startangle=30,autopct='%1.1f%%')
plt.show()

```
