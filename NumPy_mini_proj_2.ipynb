```python
import numpy as np
temp = np.random.randint(10,50, size = (30,4))

avgtemp = np.mean(temp)

maxtempdays = (np.argmax(np.max(temp, axis = 1)))
mintempdays= (np.argmin(np.min(temp,axiiis = 1 )))


z = temp[:,1]
y= z[z>35]
k = y.size


a = temp[:7]
b = np.average(a)

print(f"Average temperature of the dataset: {avgtemp}")
print(f"Hottest day of the dataset: {maxtempdays}")
print(f"Coldest day of the dataset: {mintempdays}")
print(f"Days where the afternooon temperature is more than 35 degrees: {k}")
print(f"moving average of 7days window: {b}")
```
