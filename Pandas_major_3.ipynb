```python
import pandas as pd
import matplotlib.pyplot as plt

HR = pd.read_csv("/content/WA_Fn-UseC_-HR-Employee-Attrition.csv")
h = HR.dropna()

#Department wise count
h.groupby('Department')['Department'].value_counts()

#Education Field wise count
h.groupby('EducationField')["EducationField"].value_counts()

#Maritial status with job roles
h.groupby('MaritalStatus')["JobRole"].value_counts()

#Monthly income
m = h.groupby('MonthlyIncome')["JobRole"].sum()
m.sort_values(ascending = False)

#People who work more than 10 years
mv = h.loc[h['TotalWorkingYears']>12]
mv.groupby('JobRole')["JobRole"].value_counts()
```
